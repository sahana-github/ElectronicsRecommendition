```python
file['After_Price']  = file['After_Price'].astype(str)
file['Ratings']=file['Ratings'].astype(str)
file['Title']=file['Title'].astype(str)
```

```python
file['tags'] = file['Title'] + file['After_Price'] + file['Ratings']
```

```python
file['tags'].head()
```

```python
file['tags']=file['tags'].apply(lambda x:x.lower())
```

```python
file['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(file['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[100])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(x):
    item_index=file[file['Title']==x].index[0]
    distance=similarity[item_index]
    item_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in item_list:
        print(file.iloc[i[0]].Title)
```

```python
recommend("Vivo V30 5G Smartphone, 8GB RAM 128GB Storage, Andaman Blue")
```
